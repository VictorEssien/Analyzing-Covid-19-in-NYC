```python
import pandas as pd
data = pd.read_csv("Downloads/data-by-modzcta.csv")
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
grouped_data = data.groupby("BOROUGH_GROUP")["COVID_CASE_COUNT","COVID_CASE_RATE","COVID_DEATH_RATE"]
sort = data.sort_values(by="PERCENT_POSITIVE", ascending=True)
data = sort[["BOROUGH_GROUP","PERCENT_POSITIVE","COVID_CASE_RATE","COVID_DEATH_RATE"]]
sns.set_style("whitegrid")
sns.set_context("poster",font_scale=0.6)
sns.set_color_codes("muted")
plt.subplots(figsize=(15,6))
sns.barplot(x="COVID_CASE_RATE",y="BOROUGH_GROUP",data=data,label="Total",color="red", ci=None)
plt.title("Covid-19 Case Rates in NYC")
plt.show()
```

```python
grouped_data_1 = data.groupby("BOROUGH_GROUP")["COVID_CASE_COUNT","COVID_CASE_RATE","COVID_DEATH_RATE"]
sort_1 = data.sort_values(by="PERCENT_POSITIVE", ascending=False)
data_1 = sort_1[["BOROUGH_GROUP","PERCENT_POSITIVE","COVID_CASE_RATE","COVID_DEATH_RATE"]]
sns.set_style("whitegrid")
sns.set_context("poster",font_scale=0.6)
sns.set_color_codes("muted")
plt.subplots(figsize=(15,6))
sns.barplot(x="COVID_DEATH_RATE",y="BOROUGH_GROUP",data=data,label="Total",color="red", ci=None)
plt.title("Covid-19 Death Rate in NYC")
plt.show()
```

```python
x=data["COVID_CASE_RATE"]
labels=data["NEIGHBORHOOD_NAME"]
plt.figure(figsize=(16,26))
plt.title("Covid-19 Cases in NYC (By Neighborhoods)", fontsize=20)
plt.pie(x,labels=labels)
plt.show()
```
